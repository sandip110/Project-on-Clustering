# Data Prepration

# We have to normalize the variables to express them in the range of values.

```python
# Data Normalization
from sklearn.preprocessing import normalize
scaled=normalize(df)
scaled=pd.DataFrame(scaled)
scaled.head()
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
sse=[]
for i in range(1,5):
    kmeans=KMeans(n_clusters=i, init='random', random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,5),sse)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('clustering score')
plt.show()
```

```python
sse
```

```python
# Fit the model and predict
kmeans=KMeans(n_clusters=2 , random_state=42)
kmeans.fit(df)
```

```python
pred=kmeans.predict(df)
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,pred)
```

# Hierarichal Clustering

```python
import pandas as pd 
```

```python
df=pd.read_csv(r"C:\Users\Rivigo\Desktop\Personal\Machine Learning\winequality_red.csv")
df
```

# Data Preparation

```python
from sklearn.preprocessing import normalize
scaled=normalize(df)
scaled = pd.DataFrame(scaled)
scaled.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(scaled,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Quality of wine')
plt.ylabel('apply all features')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster= AgglomerativeClustering(n_clusters=2,affinity='euclidean', linkage='ward')
#cluster.fit_predict(datasubset)
pred=cluster.fit_predict(scaled)
pred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,pred)
```

# DBSCAN Clustering

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"C:\Users\Rivigo\Desktop\Personal\Machine Learning\winequality_red.csv")
df
```

```python
cl=DBSCAN (eps=12.5, min_samples =4)
cl
```

```python
cl.fit(df)
```

```python
db=df.copy()
db
```

```python
db.loc[:,'Cluster']= cl.labels_
db
```

```python
db.Cluster.value_counts().to_frame()
```
